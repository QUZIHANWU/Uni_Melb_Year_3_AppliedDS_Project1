```python
X = dc.data_yuchuli('green_tripdata_2020-01.csv', 1, 31)
```

```python
import numpy as np
import matplotlib.pyplot as plt

feature_list = [ 'trip_distance', 'fare_amount', 'extra', 'temp']
plt.figure(figsize=(16,16))
for i in range(1,5):
    plt.subplot(220+i)
    plt.scatter(X[feature_list[i-1]], X.total_amount, c='#00CED1', alpha=0.4, label=feature_list[i-1])
    plt.xlabel(feature_list[i-1], fontsize=20)
    plt.ylabel('total_amount', size=20)
    plt.xticks(size=16)
    plt.yticks(size=16)
#     plt.xlabel(fontsize=16)
#     plt.yticks(fontsize=16)
plt.savefig('tezheng.png')
plt.show()
```
